# PyCity Schools My Analysis

•	School budget is conversely related to school performance. Schools with lower average budget/spend per student performed higher. This is evident from the Average Spend based School Analysis.

•	Charter schools perform much higher than District schools. This is evident from the Overall Passing Percentage of Charter (~95%) Vs. District (~73%) schools. 

•	District school students need more help in Math. On average there is a ballpark of 30% gap in performance of District schools Vs. Charter schools.

•	Overall performance of schools appears to be directly related to school size, with performance significantly dropping with large school sizes in the range of 2000-5000 students.

•	As a whole, current analysis is indicative of school performance, however, there is more analysis required across other factors, possibly education of teachers, student special needs etc. to prescribe areas of focus to improve District schools performance. 


```python
# Dependencies and Setup
import  os
import pandas as pd
import numpy as np

# File to Load(Remember to Change These)
schoolcsv = os.path.join("Resources","schools_complete.csv")
studentcsv = os.path.join("Resources","students_complete.csv")

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(schoolcsv)
student_data = pd.read_csv(studentcsv)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#School_data_complete.head()
#school_data_complete.dtypes
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# CAlculate District Summary
total_schools = school_data_complete['school_name'].nunique()

total_students = school_data_complete['student_name'].count()

total_budget = school_data['budget'].sum()

avg_math_score = school_data_complete['math_score'].mean()

avg_read_score = school_data_complete['reading_score'].mean()

overall_passing_rate = (avg_math_score + avg_read_score)/2

pct_math_70 = len(school_data_complete.loc[school_data_complete['math_score']>=70])/total_students

pct_reading_70 = len(school_data_complete.loc[school_data_complete['reading_score']>=70])/total_students

# Create DataFrame with District Summary

summary_df = pd.DataFrame(columns = ["Total Schools", "Total Students","Total Budget", "Average Math Score", 
                                     "Average Reading Score", "Overall Passing Rate", "Math Students(70 or Greater)",
                                     "Reading Students(70 or greater)"],
                          data =[[total_schools, total_students, total_budget, avg_math_score, avg_read_score, 
                                overall_passing_rate, pct_math_70, pct_reading_70]]
                         )
summary_df['Total Budget'] = summary_df['Total Budget'].map("${:,}".format)
           
summary_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Student ID	student_name	gender	grade	school_name	reading_score	math_score	School ID	type	size	budget
#Calculate Schoold Summary Metrics

school_summary_df = school_data_complete.groupby('school_name')


school_type = school_data.set_index('school_name')['type']

tot_students_in_school = school_summary_df['student_name'].count() 

tot_school_budget = school_summary_df['budget'].unique()

per_student_budget = tot_school_budget/tot_students_in_school

school_avg_math = school_summary_df['math_score'].mean()

school_avg_reading = school_summary_df['reading_score'].mean()

mpass_per_school =  school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name") ['student_name'].count()

rpass_per_school = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name") ['student_name'].count()

mpass_per_school = mpass_per_school/tot_students_in_school

rpass_per_school = rpass_per_school/tot_students_in_school

avg_school_score =(mpass_per_school + rpass_per_school)/2

# Capture School summary Metrics in a Dictionary

school_summary_dict = {"School Type": school_type, "Total Students": tot_students_in_school,
                      "Total School Budget": tot_school_budget, "Per Student Budget": per_student_budget,
                      "Average Math Score": school_avg_math, "Average Reading Score": school_avg_reading,
                      "% Passing Math": mpass_per_school, "% Passing Reading": rpass_per_school,
                      "Overall Passing Rate": avg_school_score}

#create a data frame with School Summary

school_summary_dict_df = pd.DataFrame(school_summary_dict)

school_summary_dict_df['Total School Budget'] = school_summary_dict_df['Total School Budget'].astype(float).map("${:,}".format)
school_summary_dict_df['Per Student Budget'] = school_summary_dict_df['Per Student Budget'].astype(float).map("${:,}".format)
school_summary_dict_df['School Type'] = school_summary_dict_df['School Type'].replace('[','')

# Top 5 Performaing Schools list

school_summary_dict_df.sort_values('Overall Passing Rate', ascending=False).head()


#school_summary_dict_df.dtypes

```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Capture average of each Grade
grade_9th= school_data_complete.loc[school_data_complete['grade'] == '9th'].groupby("school_name").mean()

grade_10th= school_data_complete.loc[school_data_complete['grade'] == '10th'].groupby("school_name").mean()

grade_11th= school_data_complete.loc[school_data_complete['grade'] == '11th'].groupby("school_name").mean()

grade_12th= school_data_complete.loc[school_data_complete['grade'] == '12th'].groupby("school_name").mean()

#Merge Grade Summary to convert Rows to Columns. This was also achievable via PivotTable, however, pivoting did not allow for 
#formatting as per instructions

score_summary1= pd.merge(grade_9th,grade_10th, on="school_name",suffixes=('_9th','_10th'))

score_summary2= pd.merge(grade_11th,grade_12th, on="school_name",suffixes=('_11th','_12th'))

score_summary= pd.merge(score_summary1,score_summary2, on="school_name")

math_score_summary = score_summary[['math_score_9th','math_score_10th','math_score_11th','math_score_12th']]


#Remove Index name as per expected format

math_score_summary.index.names = ['']

#Format Percentages

# Need to make copy of the DataFrame to avoid Panda warning "A value is trying to be set on a copy of a slice from a DataFrame."
#Copy seems to be a function with a paramater o deep. deep = True (efault), seems to be like Database Materialied View, wherein change to the original
#DataFrame are not reflected in the copied Data Frame deep=False is like a view, where changes in the original dataframe will be reflected

math_score_summary_copy = math_score_summary.copy()
math_score_summary_copy['math_score_9th']= (math_score_summary_copy['math_score_9th']/100).map('{:.2%}'.format)
math_score_summary_copy['math_score_10th']= (math_score_summary_copy['math_score_10th']/100).map('{:.2%}'.format) 
math_score_summary_copy['math_score_11th']= (math_score_summary_copy['math_score_11th']/100).map('{:.2%}'.format) 
math_score_summary_copy['math_score_12th']= (math_score_summary_copy['math_score_12th']/100).map('{:.2%}'.format)

#Rename Columns as per given format

math_score_summary_copy.rename(columns ={"math_score_9th":"9th","math_score_10th":"10th","math_score_11th":"11th",
                                    "math_score_12th":"12th" })
```
